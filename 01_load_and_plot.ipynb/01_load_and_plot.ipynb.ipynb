```python
import mne

# 1. Download (if needed) & get path to the MNE sample dataset
sample_data_folder = mne.datasets.sample.data_path()
raw_fname = f"{sample_data_folder}/MEG/sample/sample_audvis_raw.fif"

# 2. Load raw data into memory
raw = mne.io.read_raw_fif(raw_fname, preload=True)

# 3. Select the first 5 EEG channels
eeg_picks = mne.pick_types(raw.info, meg=False, eeg=True)[:5]

# 4. Plot 10 seconds of raw EEG for those channels
raw.plot(n_channels=5, picks=eeg_picks, duration=10, start=0.0)
```

```python
# Step 2: Band-pass filter between 1 and 40 Hz
raw_filtered = raw.copy().filter(l_freq=1.0, h_freq=40.0)

# Plot the filtered data to compare with the raw plot
raw_filtered.plot(n_channels=5, picks=eeg_picks, duration=10, start=0.0)

```

```python
print(raw_filtered)
```

```python
# Step 3: Artifact rejection by amplitude threshold

# 3.1 Find the stimulus events
events = mne.find_events(raw_filtered, stim_channel='STI 014')

# 3.2 Epoch the filtered data from -0.2 to 0.8 seconds around each event,
#     rejecting any epoch where EEG amplitude exceeds 100 V
epochs = mne.Epochs(
    raw_filtered,
    events,
    tmin=-0.2,
    tmax=0.8,
    preload=True,
    reject=dict(eeg=100e-6)  # units are volts, so 100e-6 = 100 V
)

# 3.3 Print how many clean epochs remain
print(f"Number of clean epochs: {len(epochs)}")

# 3.4 Plot a few of the clean epochs for visual inspection
epochs.plot(n_epochs=5, picks=eeg_picks)
```
